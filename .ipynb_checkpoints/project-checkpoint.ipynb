## Define Useful Functions

## Import Data

```python
df = pd.read_csv('customer_transaction_train.csv')
df.set_index(['ID_code'], inplace=True)
df.head()
```

For testing purposes, take randome sample of 1,000 from dataframe. Currently 200,000 samples.

```python
len(df)
```

## Clean Data

### Find Missing Values

```python
np.sum(df.isna().sum())
```

### Train Test Split

```python
X = df.drop(['target'], axis=1)
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4)
```

```python
X_train.head()
```

### Class Imbalance

Leads to the accuracy paradox

```python
y_train.value_counts()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(X_train['var_0'], X_train['var_1'], hue=y_train, palette=('RdBu'))
plt.title('Visualisation of the Class Imbalance Problem')
plt.show()
```

```python
smote = SMOTE()
X_train, y_train = smote.fit_sample(X_train, y_train)
```

```python
cc = ClusterCentroids(random_state=10)
X_train, y_train = cc.fit_resample(X_train, y_train)
```

```python
rus = RandomUnderSampler(random_state=10, replacement=True)
X_train, y_train = rus.fit_resample(X_train, y_train)
```

```python
len(X_train[:,0])
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(X_train[:,0], X_train[:,1], hue=y_train, palette=('RdBu'))
plt.title('Visualisation of the Class Imbalance Problem')
plt.show()
```

### Apply PCA to assist with Data Visualisation

```python
pca = PCA(n_components=50)
```

```python
X_pca = pca.fit_transform(X_train)
```

```python
sum(pca.explained_variance_ratio_)
```

```python
X_train_pca = pd.DataFrame(X_pca)
```

```python
fig = plt.figure(figsize=(30,20))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for n in range(1, len(X_train_pca.columns)+1):
    
    ax = fig.add_subplot(3,6,n)
    X_train_pca[[n-1]].hist(ax=ax)

fig.suptitle('Distribution of 18 Components', fontsize=40, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(X_train_pca.corr())
```

```python
X_pca_test = pca.transform(X_test)
X_test_pca = pd.DataFrame(X_pca_test)
```

###Â Identify Outliers

## K Nearest Neighbours Classifier

```python
knn = KNeighborsClassifier()
knn_param_grid = {'n_neighbors':[5,10,20,30,40,50],
              'metric': ['manhattan', 'euclidean','minkowski'],
              'weights': ['uniform', 'distance']
             }
gs_knn = GridSearchCV(knn, knn_param_grid, scoring='accuracy', cv=3)
gs_knn.fit(X_train_pca, y_train)
gs_knn.best_params_
```

```python
knn = KNeighborsClassifier(metric='manhattan', n_neighbors=20, weights='distance')
```

```python
knn.fit(X_train_pca, y_train)
```

```python
knn_preds = knn.predict(X_train_pca)
```

```python
accuracy_score(knn_preds, y_train)
```

```python
knn_preds_test = knn.predict(X_test_pca)
```

```python
accuracy_score(knn_preds_test, y_test)
```

```python
print(classification_report(knn_preds_test, y_test))
```

```python
confusion_matrix(y_test, knn_preds_test)
```

```python
fpr, tpr, thresholds = roc_curve(y_train, knn_preds)
auc(fpr, tpr)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, knn_preds_test)
auc(fpr, tpr)
```

```python
np.sum(y_train)
```

## Random Forest (Strong Learners)

```python
forest = RandomForestClassifier(class_weight='balanced')
```

```python
forest.fit(X_train_pca, y_train)
```

```python
forest_preds = forest.predict(X_train_pca)
```

```python
accuracy_score(forest_preds, y_train)
```

```python
forest_preds_test = forest.predict(X_test_pca)
```

```python
accuracy_score(forest_preds_test, y_test)
```

```python
confusion_matrix(y_test, forest_preds_test)
```

```python
fpr, tpr, thresholds = roc_curve(y_train, forest_preds)
auc(fpr, tpr)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, forest_preds_test)
auc(fpr, tpr)
```
