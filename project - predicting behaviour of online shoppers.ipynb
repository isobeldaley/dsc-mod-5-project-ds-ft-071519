# Background

Gözalan Group, a Turkish company engaged in the supply of outdoor wear, is looking to improve the conversion rate on their website.  

To assist with this task, they have asked their data science team to identify three ways in which the conversion rate can be improved, whilst managing spend.  

# The Dataset

To assist with this task Gözalan Group have provided a dataset relating to 12,330 online sessions that took place over a 1 year period.  Each record relates to a different user.  

The dataset consists of 10 numerical and 8 categorical attributes.  These attributes are:

- **Administrative**: The number of different administrative pages visited during the shopper's visit to the website
- **Administrative Duration**: The total time spent on administrative pages by the visitor during their session.
- **Informational**: The number of different informational pages visited during the shopper's visit to the website
- **Informational Duration**: The total time spent on informational pages by the visitor during their session.
- **Product**: The number of different product pages visited during the shopper's visit to the website
- **Product Duration**: The total time spent on product pages by the visitor during their session.
- **Bounce Rate**: Measured by Google Analytics.  A bounce is a single page session on a website.  Bounce rate is defined as the percentage of visitors who enter a website and then leave ("bounce") wihtout visiting any other pages.  
- **Exit Rate**: Measured by Google Analytics.  Calculated for a specific page as the proportion of times that page was the last page in the session before the customer left the site (as a percentage of all page views for that specific page).
- **Page Values**: Represents the average value for a web page that a user visited before completing an e-commerce transaction
- **Special Day**: Indicates whether the visit to the website falls on a special day such as Mother's Day, Valentine's Day etc.  
- **Month**: The month of the visit.
- **Operating System**: A numerical value indicating operating system type (non-ordinal). 
- **Browser**: A nummerical value indicating browser type (non-ordinal).
- **Region**: A numerical value indicating region (non-ordinal)
- **Traffic Type**: A numerical value indicating traffic type (non-ordinal)
- **Visitor Type**: Indicates whether this is a new or returning visitor.
- **Weekend**: Indicates whether or not it is the weekend.

In addition, there is a **Revenue** attribute, which indicates whether or not an online session culminated in an online purchase (True/False).  This will be used as the can be used as the classification (target) label.


# The Approach

To achieve the stated objective, the following approach has been employed:
- The data provided has been imported and cleaned (to identify missing values, outliers, create dummy variables etc.)
- Exploratory Data Analysis (EDA) has been performed to identify key relationships to guide model building
- The data has been split into a training and test set.  The test set will be used to ensure the model has not been overfitted, and can be generalised to new data.
- Class imbalance and the need for dimension reduction will be considered.
- A number of different classification models will be build, and performance of each will be assessed.  This includes a K Nearest Neighbour model, Random Forest, XG Boost and Support Vector Machine.
- The models will then be stacked to investigate improvements in overall performance
- The model with the best performance will be used to generate recommendations

## Useful Function

To prevent the need for repetition of code, the following function has been defined.  This function identifies the most effective combination of parameters and the best data transformation to enhance model performance.

```python
## Function to identify the optimat dataset and parameters for a given classifier and parameter grid
def best_model_parameters_dataset(classifier, param_grid, datasets):
    
    ## Create a list to contain the dataset, optimal parameters, and score for training and test set
    score_parameters = []
    
    ## Create a for loop which iterates through each dataset and identifies the optimal parameters for the given classifier
    for data in datasets:
        
        gs = GridSearchCV(classifier, param_grid, scoring='accuracy', cv=3)
        gs.fit(data['X_train'], data['y_train']['Revenue_True'])
        y_test_preds = gs.predict(data['X_test'])
        test_score = accuracy_score(y_test_preds, data['y_test']['Revenue_True'])
        score_parameters.append({'Dataset':data['name'], 'Training Score':round(gs.best_score_,2), 'Test Score': round(test_score,2), 'Parameters':gs.best_params_})
     
    ## Generate a dataframe that contains the optimal parameters for each dataset
    df = pd.DataFrame(score_parameters)
    df.sort_values(by=['Test Score', 'Training Score'], inplace=True, ascending=False)
    
    return df
```

### Check for Class Imbalance

A common issue with classification models is the presence of class imbalance.  Class imbalance arises when the total number observations assigned to one class is far greater than the number of observations assigned to another class.

In this case, class imbalance will be present if the number of observations in which the customer makes/does not make a purchase is far greater than the number of observations in which the customer does not make/makes a purchase.  

To check for class imabalnce, the **value_counts()** function can be deployed on the target field (**Revenue_True**).

```python
## Apply the value_counts() function to identify the number of observations that fall into each class
y_train['Revenue_True'].value_counts()
```

As can be seen above, there is a class balance problem.  The number of observations in which a customer makes a purchase is 1625 versus 9756 observations in which a purchase is not made.  

In practice, this means that if we were to create a baseline model that predicted every value to be 0, it would have an accuracy of 86% (1625/(1625+9756)).  This means, in order for a more sophisticated mode to add any value, it must have an accuracy whihc exceeds 86%.

Class imbalance can adversely affect the performance of a classification model.  One way of addressing this issue is through oversampling.  To apply this procedure, Synthetic Minority Oversampling Technique (SMOTE) will be used.  SMOTE works by generating new sample data for the minority class by creating 'syntetic' samples that are combinations of the closest minority class observations.  

```python
## Use the SMOTE() function to create 'synthetic' samples for the minority class
smote = SMOTE()
X_train_smote, y_train_smote = smote.fit_sample(X_train, y_train)

## Push X_train_smote and y_train_smote into a Pandas dataframe (from a Numpy array)
X_train_smote = pd.DataFrame(X_train_smote, columns=X.columns)
y_train_smote = pd.DataFrame(y_train_smote, columns=y.columns)
```

### Apply Principal Component Analysis (PCA)

Our current dataset has a total of 64 features.  It is possible that this high dimensionality will lead to some problems.  In a 64-dimension space, there is likely to be a significant amount of "white space" (i.e. areas in which we have no observations).  This could compromise the accuracy of the models.  

To address this issue, Principal Component Analysis (PCA) can be applied.  PCA attempts to capture as much information (variance) as possible, whilst also reducing the number of dimensions in the dataset.  

Scikit-learn contains the functionality required to perform PCA.  PCA will be applied to create two separate datasets:
- The first will have applied PCA to the original dataset 
- The second will have applied PCA to the balanced (SMOTE) dataset

```python
## Apply PCA to the original (unbalanced) dataset

## Create counter variables to identify the optimal number of components to use
## in order to explain at least 99% of the variance

components = 0
explained_variance = 0

## Use a while loop to incrementally add a component, and calculate the explained variance for
## this number of components.  Once 99% of variance is explained, stop the loop
while explained_variance < 0.99:
    components+=1
    pca = PCA(n_components=components)
    X_train_pca = pca.fit_transform(X_train)
    explained_variance = sum(pca.explained_variance_ratio_)
    

## Print the number of PCA components used, and the amount of variance explained
print("Use ",components," for PCA Dimension Reduction in the original (unbalanced) dataset")
print("This explains ", round(explained_variance*100,2), "% of the variance in the original (unbalanced) dataset" )
```

The PCA transformation must also be applied to the test datasets.  Note that the transform applies the eigenvectors/eigenvalues generated on the training dataset.  This ensures that the same features are used.

```python
## Apply the transformation to the test data for the original (unbalanced) dataset 
X_test_pca = pca.transform(X_test)
```

```python
## Apply PCA to the balanced (SMOTE) dataset

## Create counter variables to identify the optimal number of components to use
## in order to explain at least 99% of the variance

components = 0
explained_variance = 0

## Use a while loop to incrementally add a component, and calculate the explained variance for
## this number of components.  Once 99% of variance is explained, stop the loop
while explained_variance < 0.99:
    components+=1
    pca = PCA(n_components=components)
    X_train_smote_pca = pca.fit_transform(X_train_smote)
    explained_variance = sum(pca.explained_variance_ratio_)
    

## Print the number of PCA components used, and the amount of variance explained
print("Use ",components," for PCA Dimension Reduction in the SMOTE dataset")
print("This explains ", round(explained_variance*100,2), "% of the variance in the SMOTE dataset" )
```

The PCA transformation must also be applied to the test datasets.  Note that the transform applies the eigenvectors/eigenvalues generated on the training dataset.  This ensures that the same features are used.

```python
## Apply the transformation to the test data for the original (unbalanced) dataset 
X_test_smote_pca = pca.transform(X_test)
```

#### Classification Report

The classification report provides the following key metrics:
- **Precision**: Defined as the number of times the model correctly assigned a given classification,  as a proportion of all observations with that predicted classification.  A low precison suggests a high rate of false positives.  
- **Recall**: Defined as the number of times the model correctly assigned a given classification,  as a proportion of all observations with that actually had that classification.  A low recall suggests a high rate of false negatives.  
- **f1-score**: The harmonic mean of precision and recall.  
- **Accuracy**: The proportion of all observations that were correctly classified

Note that there is an inverse relationship between precision and recall.  Balance is therefore important.  

```python
## Print classification report for predictions generated by training data
print(classification_report(y_train,knn_preds_train))
```

```python
## Print classification report for predictions generated by test data
print(classification_report(y_test, knn_preds_test, ))
```

- There is a good balance in accuracy between the training and the test model (91% v. 90%).  This indicates that the model has not been overfitted and can be generalised to new data.



- There is a good balance between precision and recall for both thee training ad test models when predicting customers that did not purchase (0). ## Stacking the Models

To see if the performance of the models can be improved further, the classification algorithms will be stacked in every possible combination.  The motivation behind doing this can be found in the concept of *wisdom of the crowd*.  This principle asserts that the collective opinion of a group (in our case a group of models) will produce more insight than the opinion of a single entity.  

The models can be stacked by taking the predictions from each individual model as an input, and then running a new classification model.  

Before stacking, the performance of each model is ranked by their AUC Test score.

```python
## Print AUC for each model type for test data
auc_test = pd.DataFrame([knn_auc_test, logreg_auc_test, forest_auc_test, boost_auc_test, svm_auc_test], 
                       index=['K Nearest Neighbours', 'Logistic Regression','Random Forest','XG Boost', 'SVM'],
                       columns=['AUC Test Score'])
auc_test.sort_values(by=['AUC Test Score'], inplace=True, ascending=False)
auc_test
```

As can be seen from the above, the Logistic Regression performs best.  All of the other models, with the exception of the Support Vector Machine, also have good performance according to this metric.  

Before stacking the models, a two DataFrames containing the training and test predictions are compiled.  These will be used as the inputs to the new classifier.  

```python
## Create a dataframe with the predictions generated using the training set
X_train_stacked = pd.DataFrame({'knn': knn_preds_train, 'logreg': logreg_preds_train, 
                                'boost': boost_preds_train, 'forest':forest_preds_train,
                               'svm':svm_preds_train})

## Create a dataframe with the predictions generated using the test set
X_test_stacked = pd.DataFrame({'knn': knn_preds_test, 'logreg': logreg_preds_test, 
                                'boost': boost_preds_test, 'forest':forest_preds_test,
                               'svm':svm_preds_test})
```

Next, a classifier is chosen.  In this case, we will use the K Nearest Neighbours classifier.

```python
## Define the classifier to be used
stacked_clf = KNeighborsClassifier()
```

Finally, to ensure thorough assessment, all possible combinations of stacking the models will be explored.  This is achieved by creating a list of all possible combinations (and lenghts), and then estimating the model and AUC score.

```python
## Define the names of all models to be assessed
models = ['knn','logreg', 'boost','forest','svm']

## Create an empty list to contain the AUC scores for each model combinations
auc_stacked = []

## Create a for loop which runs through each possible combination length (2,3,4 or 5)
for i in range(2,6):
    
    combos = list(combinations(models,i))
    
    ## For each set of combinations, run the model
    for key,val in enumerate(combos):
        
        models_to_include = []
        
        for k in range(0,len(combos[0])):
            
            models_to_include.append(combos[key][k])
        
        stacked_clf.fit(X_train_stacked[models_to_include], y_train['Revenue_True'])
        
        stacked_preds_train = stacked_clf.predict(X_train_stacked[models_to_include])
        stacked_preds_test = stacked_clf.predict(X_test_stacked[models_to_include])
        
        ## Caclulate the AUC score for each model

        fpr_train, tpr_train, thresholds_train = roc_curve(y_train['Revenue_True'], stacked_preds_train)
        auc_train = round(auc(fpr_train, tpr_train),2)

        fpr_test, tpr_test, thresholds_test = roc_curve(y_test['Revenue_True'], stacked_preds_test)
        auc_test = round(auc(fpr_test, tpr_test),2)

        auc_stacked.append({'Model Name': val, 'AUC Train': auc_train, 'AUC Test':auc_test})  
    
```

The AUC scores for each model are consolidated into a DataFrame.

```python
## Consolidate auc_stacked into a dataframe
auc_stacked_df = pd.DataFrame(auc_stacked)
auc_stacked_df.sort_values(by=['AUC Test', 'AUC Train'], inplace=True, ascending=False)
auc_stacked_df.head()
```

As can be seen from the above, the highest performing stacked model uses the Logistic Regression and XG Boost.  However, none of these models perform as well as the Logistic Regression on its own.  

For this reason, the Logistic Regression model will be chosen to proceed.

## Interpreting the Model & Making Recommendations

### Understanding the Importance of Each Feature

To evaluate the importance of each predictor in determining a customer's propensity to purchase, Recursive Feature Elimination (RFE) will be used.  RFE recusively fits a model, removing the weakest feature each time.  It can therefore be employed to isolate the most important determinants of behaviour in this model.

```python
## Run RFE on the Logistic Regression model
rfe = RFE(logreg, 1)
rfe = rfe.fit(X_train, y_train)

## Create a dataframe containing the coefficients of the Logistic Regression, 
## and their ranking according to RFE
parameter_info = {'Coefficients':logreg.coef_.tolist()[0], 'Ranking': rfe.ranking_.tolist()}
parameter_df = pd.DataFrame(parameter_info, index=list(X_train.columns))

## Calculate the Log Odds ratio, and round this and the coefficients value to 2 decimal places
parameter_df['Log Odds Ratio'] = round(np.exp(parameter_df['Coefficients']),2)
parameter_df['Coefficients'] = round(parameter_df['Coefficients'],2)

## Sort the features by ranking (i.e. show the most important first)
parameter_df.sort_values(by=['Ranking']).head(10)
```

From the above, we can see that the most important features are **ExitRates** and **BounceRates**.  This is followed by **TrafficType_15** and **TrafficType_18** and **Month_Feb**.  All of these features have a negative impact on the likelihood of a consumer to purchase.

### Visualising the Effect of the Ordinal Variables

To visualise the effects of the ordinal variables, a probability plot will be created.  The probability plot shows the value of a given independent variable, against the estimated probability of a customer making a purchase.  

The ordinal variables examined are:
- **BounceRates**
- **ExitRates**
- **PageValues**
- **SpecialDay**
- **Total_Duration**
- **Total_Pages**

```python
## Create a figure with 6 subplots, one per independent variable
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(20,30))

## Run a for loop, in which a subplot is created for each independent variable
for key, coefficient in enumerate(list(parameter_df.index[0:6])):
    
    ## Specify the subplot on which the given independent variable is to be plotted
    i = key//2
    j = key%2
    
    ## plot the independent variable against the estimated probability
    sns.regplot(X_test[coefficient], logreg.predict_proba(X_test)[:,1], ax=axes[i][j])
    axes[i][j].set_ylabel('Probability of Purchase')
    axes[i][j].set_ylim([0,1.01])
    axes[i][j].set_xlim([0, X_test[coefficient].max()+0.1])
    axes[i][j].set_yticks(np.arange(0,1,0.05))
    axes[i][j].set_title(f"""Probability Plot: {coefficient}""")
    
    
plt.show()
```

From the above, we can draw the following conclusions:
- There is a notable negative relationship between bounce rate/exit rate and the probability of a customer making a purchase.  This seems logical - if a given customer leaves the website after visitng one page, they are far less likely to place an order.
- There is a positive relationship between PageValues, Total_Durationa dn Total_Pages and the probability of a customer placing an order.  It follows that the longer a customer spends on a particular site, and the more pages they visit, the more committed they are to making a purchase.  
- There is a negative relationship between SpecialDay and the likelihood of a purchase.  SpecialDay indicates the closeness of the visit date to a special day (e.g. Mothers' Day, Valentine's Day etc.).  The negative relationship suggests the closer one is to a Special Day, the less likely a customer is to purchase.  Perhaps this is because they are planning ahead.
